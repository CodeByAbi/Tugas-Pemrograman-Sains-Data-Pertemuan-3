# **Hitung nilai akhir (30% tugas, 30% UTS, 40% UAS)**

```python
#Menambahkan nilai Tugas, UTS, UAS
data = {'Name' : ["Abira Wisnunggal", "Nouval Ghonimu Tsawab", "Muhammad Lintang Saudjana", "Zainab Amri"],
        'Gender' : ["Laki-Laki", "Laki-Laki", "Laki-Laki", "Perempuan"],
        'No. Telepon' : ["081267001022", "087823560918", "089654009102", "082267018967"],
        'Tugas' : [80,89,76,80],
        'UTS' : [88,78,90,81],
        'UAS' : [90,89,85,82]}

df = pd.DataFrame(data)
df

df["Nilai Akhir"] = 0.3 * df['Tugas'] + 0.3 * df['UTS'] + 0.4 * df['UAS']
df

```
